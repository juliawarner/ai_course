```
import numpy as np
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

# Local Minimum

```
#define 3d function to graph
def f1(x, y):
    return x ** 2 + y ** 2

#generate x and y values
x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

#create meshgrid from x and y values
X, Y = np.meshgrid(x, y)

#generate Z meshgrid values
Z = f1(X, Y)

#plot 3d graph
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary', zorder = -1000)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

#mark point at (0, 0, 0), a local minimum
ax.plot3D([0], [0], [0], markerfacecolor='r', markeredgecolor='r', marker='.', markersize=20, alpha=1.0, zorder = 1000);
```

# Local Maximum

```
#define 3d function to graph
def f1(x, y):
    return -(x ** 2) - (y ** 2)

#generate x and y values
x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

#create meshgrid from x and y values
X, Y = np.meshgrid(x, y)

#generate Z meshgrid values
Z = f1(X, Y)

#plot 3d graph
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary', zorder = -1000)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

#mark point at (0, 0, 0), a local maximum
ax.plot3D([0], [0], [0], markerfacecolor='r', markeredgecolor='r', marker='.', markersize=20, alpha=1.0, zorder = 1000);
```

# Zero Gradient, not min or max

```
#define 3d function to graph
def f1(x, y):
    return 0 - x ** 3 - y ** 2

#generate x and y values
x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

#create meshgrid from x and y values
X, Y = np.meshgrid(x, y)

#generate Z meshgrid values
Z = f1(X, Y)

#plot 3d graph
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary', zorder = -1000)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

#mark point at (0, 0, 0), neither a minimum nor maximum, but the gradient is zero at this point
ax.plot3D([0], [0], [0], markerfacecolor='r', markeredgecolor='r', marker='.', markersize=20, alpha=1.0, zorder = 1000);
```
