# Linear regression using numpy


```
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rc('axes', labelsize=14)
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)
```

```
import numpy as np
np.random.seed(42)

xs = 2 * np.random.rand(100, 1)
ys = 4 + 3 * xs + np.random.rand(100, 1)
```

```
plt.plot(xs, ys, "b.")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.axis([0, 2, 0, 15])
plt.show()
```

```
# split the data into training and test sets
# train set
train_xs = xs[:80]
train_ys = ys[:80]
# test set
test_xs = xs[80:]
test_ys = ys[80:]
```

```
# number of epochs
epochs = 10
# learning rate
lr = 0.01
```

```
# initial value for weight w and bias b
w = np.random.randn(1)
b = np.zeros(1)
```

```
for epoch in np.arange(epochs):
  for i in np.arange(80):
    y_pred = w * train_xs[i] + b
    
    grad_w = (y_pred - train_ys[i]) * train_xs[i]
    grad_b = (y_pred - train_ys[i])
    
    w -= lr * grad_w
    b -= lr * grad_b
```

```
test_loss = 0
for i in np.arange(20):
  test_loss += 0.5 * (w * test_xs[i] + b - test_ys[i]) ** 2
test_loss /= 20
```

```
test_loss
```

```
pred_ys = w * test_xs + b
```

```
plt.plot(test_xs, test_ys, "b.")
plt.plot(test_xs, pred_ys, "r.") # predicted values
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.axis([0, 2, 0, 15])
plt.show()
```

# Linear regression with two weights and one bias term

```
from mpl_toolkits import mplot3d
```

```
#declare actual variable values
w1_ground_truth = 4
w2_ground_truth = 6
b_ground_truth = 4

#generate data
np.random.seed(42)
xs = 2 * np.random.rand(100, 1)
ys = w1_ground_truth * xs + np.random.rand(100, 1)
zs = w2_ground_truth * xs + b_ground_truth + np.random.rand(100, 1)
```

```
#plot the data
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.scatter(xs, ys, zs, "b.");
```

```
#split the data into training and testing sets
# train set
train_xs = xs[:80]
train_ys = ys[:80]
train_zs = zs[:80]
# test set
test_xs = xs[80:]
test_ys = ys[80:]
test_zs = zs[80:]
```

```
# number of epochs
epochs = 10
# learning rate
lr = 0.01
```

```
# initial value for weights w1, w2 and bias b
w1 = np.random.randn(1)
w2 = np.random.randn(1)
b = np.zeros(1)
```

```
#train the model
for epoch in np.arange(epochs):
  for i in np.arange(80):
    y_pred = w1 * train_xs[i]
    z_pred = w2 * train_xs[i] + b
    
    grad_w1 = (y_pred - train_ys[i]) * train_xs[i]
    grad_w2 = (z_pred - train_zs[i]) * train_xs[i]
    grad_b = (z_pred - train_zs[i])
    
    w1 -= lr * grad_w1
    w2 -= lr * grad_w2
    b -= lr * grad_b
```

```
#compare loss at each point in testing data, find average loss
test_loss = 0
for i in np.arange(20):
  test_loss += 0.5 * ((w1 * test_xs[i] - test_ys[i]) + (w2 * test_xs[i] + b - test_zs[i])) ** 2
test_loss /= 20

print('Average loss is ' + str(test_loss))
```

```
#plot predictions vs ground truth for the testing data
pred_ys = w1 * test_xs
pred_zs = w2 * test_xs + b
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.scatter(test_xs, test_ys, test_zs, c = "b")
ax.scatter(test_xs, pred_ys, pred_zs, c = "r");
```

```
# print calculated weights and bias term
print('Predicted weight 1 was ' + str(w1) + ', actual weight 1 was ' + str(w1_ground_truth))
print('Predicted weight 2 was ' + str(w2) + ', actual weight 2 was ' + str(w2_ground_truth))
print('Predicted bias term was ' + str(b) + ', actual bias term was ' + str(b_ground_truth))
```
